```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
from sklearn import preprocessing 
```

```python
iris = pd.read_csv('iris_dataset.csv')
iris
```

```python
plt.scatter(iris['sepal_length'], iris['sepal_width'])
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')
plt.show()
```

```python
x = iris.copy()
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
iris['Cluster'] = kmeans.fit_predict(x)
iris
```

```python
plt.scatter(iris['sepal_length'], iris['sepal_width'], c = iris['Cluster'], cmap = 'rainbow')
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')
plt.show()
```

```python
#standardize the data
x_standard = preprocessing.scale(x)
x_standard
```

```python
kmeans_standard = KMeans(2)
kmeans_standard.fit(x_standard)
standard_data = iris.copy()
standard_data['Cluster'] = kmeans_standard.fit_predict(x_standard)
standard_data
```

```python
plt.scatter(standard_data['sepal_length'], standard_data['sepal_width'], c = standard_data['Cluster'], cmap = 'rainbow')
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')
plt.show()
```

```python
wcss = []

for i in range(1, 10):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
wcss
```

```python
#Elbow method
plt.plot(range(1, 10), wcss)
plt.xlabel("Number of clusters")
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans_standard = KMeans(3)
kmeans_standard.fit(x_standard)
standard_data['Cluster'] = kmeans_standard.fit_predict(x_standard)

plt.scatter(standard_data['sepal_length'], standard_data['sepal_width'], c = standard_data['Cluster'], cmap = 'rainbow')
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')
plt.show()
```
